# ¿Qué es Matplotlib?

- **Matplotlib** es una biblioteca para la creación de visualizaciones tanto estáticas, animadas e interactivas en Python. Permite: 
    - Creción de gráficos de alta calidad.
    - Figuras interactivas a las que hacer zoom, efectos panorámicos y actualizaciones.
    - Personalizar estilos y diseños
    - Exportación a una amplia gama de formatos de archivos.
    - Compatible con JupyterLab/Jupyter Notebook e interfaces de usuario.
    - Contiene una amplia cantidad de paquetes de terceros desarrollados en Matplotlib
- La biblioteca Pandas contiene wrappers (envoltorios) en sus series y dataframes para la generación de gráficos.
-  Si está trabajando en su entorno local, previo a su importación debe ser **instalada** 
    - ```pip install matplotlib``` si usa ```pip```
    - ```conda install matplotlib``` si usa ```conda```

````python
import pandas as pd
import matplotlib.pyplot as plt

fig1 = plt.figure(dpi=70)
ax1 = fig1.add_subplot(1,2,1)
ax1.plot(([1,2,3,4]), [1,4,9,16], 'o--', color="orchid", markerfacecolor="lightgreen", lw=2, ms=10)
ax1.axis([0,5,0,17])

ax2 = fig1.add_subplot(1,2,2)
ax2.plot(([1,2,3,4]), [1,4,9,16], 'o--', color="red", markerfacecolor="yellow", lw=2, ms=10)
ax2.axis([0,5,0,17])

plt.show()
````

# Utilizando texto en gráfico

````python
from matplotlib import pyplot as pyplot

fig = plt.figure(dpi=95)
fig.suptitle("Titulo superior en negrita (bold)", fontsize=14, fontweight="bold")

ax= fig.add_subplot(1,1,1)
fig.subplots_adjust(top=1) #ajustar posicion del gráfico

ax.set_title("Titulo de ejes, x vs y")
ax.set_xlabel("Etiqueta de eje x")
ax.set_ylabel("Etiqueta de eje y")

#TEXO A COLOR
ax.text(1.5, 8, 'Texto a todo color en coordenadas (1.5, 8)', color="darkgreen", fontsize=12)

#GRAFICAMOS UN PENTAGONO Y UNA LISTA SEGMENTADA
ax.plot([2],[6], 'p', ms=10)
ax.plot([1,2,3,4,5,6,7,8], [0.6,3,2,1,4,5,6,1], 'rd--', lw=2)

# GENERAMOS ANOTACION CON UNA FLECHA: CORDENADA (2,1), TEXTO DE FLECHA COORDENADA (3,1)
ax.annotate('Este es  un Outlayer', xy=(2,6), xytext=(4,7), arrowprops=dict(facecolor='pink', shrink=0.02))

ax.axis([0,10,0,10])
plt.show()
````

````python
#LOS DATOS DE TORTA ES LA VIEJA CONFIABLE
# SE ANALIZA MUY BIEN POR QUE NOS PERMITE A NOSOTROS VER PROPORCIONES
# ES MUY RELEVANTE Y POR LO GENERAL ES UNIVARIABLE

import matplotlib.pyplot as plt

#GRÁFICO DE TORTA; CORTES ORDENAODS Y GRÁFICADOS CONTRA RELOJ

size = [15,40,60,20]
equipos = ["Alianza","Real Madrid", "Celta", "Perú"]
explotar = [0,0,0.1,0] # sirve para destacar alguna porcion, separación

fig3 = plt.figure(dpi=90)
ax1 = fig3.add_subplot(1,1,1)

ax1.pie(size, labels=equipos, explode=explotar, autopct="%.2f%%", shadow=False, startangle=-90)

ax1.axis("equal")

plt.show()

````

## Ahora utilizaremos el dataframe de Mortgages

````python
df = pd.read_csv("data/mortgages.csv")
df.head()
````

````python
df_aux = df.groupby(["Month","Interest Rate",])[["Interest Paid"]].sum().reset_index() 
df_aux
````

## Se puede observar que tenemos datos repetidos, tenemos que filtrar datos para que  nos quede lo mas limpio posible a la hoara de realizar las gráficas

````python
##filtros
filtro_1 = df_aux["Interest Rate"] == 0.03
filtro_2 = df_aux["Interest Rate"] == 0.05

df_aux_1 = df_aux[filtro_1]
df_aux_2 = df_aux[filtro_2]

df_aux_1
````

````python
plt.plot(df_aux_1["Month"], df_aux_1["Interest Paid"], label="Interes de 3%")
plt.plot(df_aux_2["Month"], df_aux_2["Interest Paid"], label="Interes de 5%")
plt.legend()
plt.show()
````
