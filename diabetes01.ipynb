```python
import pandas as pd
```

```python
df = pd.read_csv("data/diabetes.csv")
df.head()
```

```python
df.info()
```

```python
#funcion estadística que nos ayuda analizar las columnas numericas 
#Q3 = 6
#Q2 = 3
#Q1 = 1
#Entre Q3 y Q1 = 5
# 5 + Q3 = 11 -> valores por encima de 11 son considerados outliers
df.describe()
```

```python
# Primer filtro cuando hay mas de 11 embarazos
filtro_1 = df["Pregnancies"]>=11
df_2=df[filtro_1]
df_2.head()
```

```python
df_2.info()
```

```python
# Observamos que la insulina tiene valores extraños
df_2.describe()
```

```python
#Eliminamos insulima de este dataset, se pudo haber dañado
del df_2["Insulin"]
df_2.head()
```

```python
# guardamos este dataset en un excel
df_2.to_excel("dataframe_2_pregnancies.xlsx", index=False)
```

```python
# Se observó en el excel que hay valores demasiados extraños en la columna DiabetesPedigreeFunction

filtro_2 = df["BloodPressure"] == 0
df_3 = df[filtro_2]
df_3.head()
# Al parecer estos valores tienen muchos problemas(ceros)
```

```python
df_3.info()
```

```python
df_3.describe()
```

```python
# Observamos que la media de insulin es 0 y toda DiabetesPedigreeFunction es cero 
# entonces eliminamos esas columas 

df_3.drop(columns=["Insulin","DiabetesPedigreeFunction"], inplace=True)
df_3.head()
```
